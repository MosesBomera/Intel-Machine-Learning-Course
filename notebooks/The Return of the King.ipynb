```python
import numpy as np
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('Simple Graph')
```

```python
#f, (ax1, ax2) = plt.subplots(1,2, sharey=True)
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```
